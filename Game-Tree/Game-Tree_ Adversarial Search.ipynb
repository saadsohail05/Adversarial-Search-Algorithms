# Alpha-Beta Pruning - Game Tree 3
![Game Tree 3.png](<attachment:Game Tree 3.png>)

```python
def alphabetapruning(graph, nodes, startnode):
    visitednodes = [0]
    maxdepth = [0]

    bestval, bestpath = maxvalue(graph, nodes, startnode, [], float('-inf'), float('inf'), visitednodes, maxdepth)
    print("AlphaBeta: Utility Value of Root Node: =", bestval)
    print("Best Path:", bestpath)
    print("Time Complexity - Number of nodes generated:", visitednodes[0])
    print("Space Complexity - Maximum Depth:", maxdepth[0])

def maxvalue(graph, nodes, node, path, alpha, beta, visitednodes, maxdepth):
    visitednodes[0] += 1
    maxdepth[0] = max(maxdepth[0], len(path) + 1)

    if not isterminal(graph, node):
        print(f"Visited Node: {node} -> Max")
    else:
        print(f"Visited Node: {node}")

    if isterminal(graph, node):
        return getutility(nodes, node), path + [node]

    maxval = float('-inf')
    bestpath = []

    for state in getsuccessors(graph, node):
        value, subpath = minvalue(graph, nodes, state, path + [node], alpha, beta, visitednodes, maxdepth)
        if value > maxval:
            maxval = value
            bestpath = subpath
        alpha = max(alpha, maxval)
        if alpha >= beta:
            break  
    return maxval, bestpath

def minvalue(graph, nodes, node, path, alpha, beta, visitednodes, maxdepth):
    visitednodes[0] += 1
    maxdepth[0] = max(maxdepth[0], len(path) + 1)

    if not isterminal(graph, node):
        print(f"Visited Node: {node} -> Min")
    else:
        print(f"Visited Node: {node}")

    if isterminal(graph, node):
        return getutility(nodes, node), path + [node]

    minvalue = float('inf')
    bestpath = []

    for state in getsuccessors(graph, node):
        value, subpath = maxvalue(graph, nodes, state, path + [node], alpha, beta, visitednodes, maxdepth)
        if value < minvalue:
            minvalue = value
            bestpath = subpath
        beta = min(beta, minvalue)
        if alpha >= beta:
            break  
    return minvalue, bestpath

def getsuccessors(graph, node):
    if node in graph:
        return graph[node]
    else:
        print("Node not found in the graph")
        return []

def isterminal(graph, node):
    if node in graph:
        return len(graph[node]) == 0
    else:
        print("Node not found in the graph")
        return False

def getutility(nodes, node):
    return nodes.get(node)

def main():
   graph = {
    'A': ['B', 'C', 'D'],
    'B': ['3', '12', '8'],
    'C': ['2', 'x', 'y'],
    'D': ['14', '5', '2'],
    '3': [],
    '12': [],
    '8': [],
    '2': [],
    'x': [],
    'y': [],
    '14': [],
    '5': [],
    '2': []
}

   leafnodes = {
    '3': 3,
    '12': 12,
    '8': 8,
    '2': 2,
    'x': 0,
    'y': 0,
    '14': 14,
    '5': 5,
    '2': 2
}

   alphabetapruning(graph, leafnodes, "A")

main()
```
